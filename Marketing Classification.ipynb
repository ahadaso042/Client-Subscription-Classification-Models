```python
train_data["Subscribed_val"] = train_data["Subscribed"]

train_data["Subscribed_val"] = train_data["Subscribed_val"].map({'yes': 1, 'no': 0})
#Drop Subscribed column
```

```python
test_data["Subscribed_val"] = test_data["Subscribed"]

test_data["Subscribed_val"] = test_data["Subscribed_val"].map({'yes': 1, 'no': 0})
#Drop Subscribed column
```

```python
train_data = pd.get_dummies(train_data, columns=["job", "marital", "education", "housing", "loan", "contact", "month", "day_of_week", "poutcome"])
```

```python
test_data = pd.get_dummies(test_data, columns=["job", "marital", "education", "housing", "loan", "contact", "month", "day_of_week", "poutcome"])
```

```python
X = train_data_num.iloc[:,0:5]
Y = train_data_num.iloc[:,-1]

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model = ExtraTreesClassifier()
model.fit(X,Y)
print(model.feature_importances_)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn import tree
tree1 = tree.DecisionTreeClassifier()
tree1 = tree1.fit(train_data[['duration','age','pdays', 'poutcome_success','poutcome_failure']], train_data['Subscribed_val'])
```

```python
predictions = tree1.predict(test_data[['duration','age','pdays', 
                                    'poutcome_failure','poutcome_success']])
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_data[['Subscribed_val']], predictions))
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(X=train_data[['duration','age','pdays', 'poutcome_success','poutcome_failure']], y=train_data['Subscribed_val'])
```

```python
import matplotlib.pyplot as plt

m = model1.coef_
b = model1.intercept_

print('slope of model 1:', m)
print('intercept of model 1:', b)
```

```python
predicted_val1 = model1.predict(test_data[['duration','age','pdays', 'poutcome_success','poutcome_failure']])
print('predicted values using model 1:', predicted_val1)

real_values = test_data['Subscribed_val'].to_numpy()
print('real values:', real_values)

diff1_array = predicted_val1 - real_values

diff1_array = np.absolute(diff1_array)
diff1_sum = np.sum(diff1_array)

print('the sum of the differences between real and predicted values for model 1 is:', diff1_sum)

print("Mean Absolute Error (MAE) for model 1 is: ", diff1_sum/len(diff1_array))
```
